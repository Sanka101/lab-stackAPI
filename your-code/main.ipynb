#### Question 1: Find the questions and answers of last month.

```python
import time
import datetime
# Time stamp is a way to track time as a running total of seconds.
# This count starts at the Unix Epoch on January 1st, 1970.
```

```python
# timestamp
# from 6th of June to 6th of July "set" dates
date1 = '06/06/2020'
date2 = '06/07/2020'
# make timestamp
ts1 = time.mktime(time.strptime(date1, "%d/%m/%Y"))
ts2 = time.mktime(time.strptime(date2, "%d/%m/%Y"))
print(ts2)
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# question, tagged and score 5

questions = SITE.fetch('questions', fromdate=1591394400, todate=1593986400, 
                      tagged= 'python', sort='votes', min=5)
```

```python
pd.DataFrame(questions)
print(questions)
```
